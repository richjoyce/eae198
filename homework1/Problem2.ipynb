```python
# Set up notebook environment
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
from IPython.display import HTML
# Set up some constants
r_e = 6378.1       # km; radius of the earth
mu = 398600.4418   # km^3/s^2; gravitational parameter of earth
```

# Problem 2. Orbital Maneuvers

#### (a) Using the orbital maneuver formulas, calculate the values shown in Table 9-18 in SMAD.

```python
from IPython.display import Image
display(Image("../../resources/images/table918.png"))
```

For this problem we are being asked to determine the relative amount of $\Delta V$ use for varying orbit transfer strategies.
The key here is that the starting orbit and finish orbit are the exact same.
In fact, since we are only doing Hohmann transfers, the elliptical transfer size and shape will be the same, only the inclination will change with each strategy.

First, let's define the lower and upper orbits in some variables.

```python
r_circ = 250.0 + r_e        # Radius of the lower circular orbit (km)
i_circ = np.deg2rad(28.5)   # Inclination of the circular orbit (rad)
r_geo  = 35786.0 + r_e       # Radius of the geostationary orbit (km)
i_geo  = 0.0                # Inclination of the circular orbit (rad)
```

We are going to make heavy use of SMAD 9-68 through 9-70, 9-77 and 9-78.
SMAD 9-68 describes the semimajor axis of the Hohmann transfer orbit.
$a_l$ and $a_h$ are the altitudes of the lower and higher orbit, respectively.
$$a_t = \frac{a_l+a_h}{2}$$
In order to complete a Hohmann transfer, we need to complete two burns, one to enter our transfer orbit from the low earth orbit circular orbit, the second to circularize to the GEO orbit, at the apogee of the transfer orbit.
The magnitude of these two burns are given by 9-69 and 9-70.
$$\Delta V_H = \sqrt{\mu} \left( \sqrt{ \frac{2}{a_l} - \frac{1}{a_t} } - \sqrt{ \frac{1}{a_l} } \right)$$
$$\Delta V_L = \sqrt{\mu} \left( \sqrt{ \frac{2}{a_l} - \frac{1}{a_t} } - \sqrt{ \frac{1}{a_l} } \right)$$
Finally, we need to determine how much $\Delta V$ a plane change will use.
A simple plane change is described by SMAD 9-78.
$$\Delta V = 2V\sin\left(\frac{\theta}{2}\right)$$
This is a special case of SMAD 9-77, which is the combined plane change with a $\Delta V$ along the orbit direction too.
You can see this reduces to the previous formula if $V_i=V_f$.
$$\Delta V = \sqrt{V_i^2 + V_f^2 - 2V_iV_f\cos(\theta)}$$
Now let's write up our equations as functions.

```python
def hohmann(a_l,a_h):
    global mu
    a_t = 0.5*(a_l+a_h) # SMAD 9-68
    dV_l = np.sqrt(mu)*(np.sqrt(2.0/a_l-1/a_t)-np.sqrt(1.0/a_l)) # SMAD 9-69
    dV_h = np.sqrt(mu)*(np.sqrt(1.0/a_h)-np.sqrt(2.0/a_h-1/a_t)) # SMAD 9-70
    return dV_l,dV_h

def plane_change(a, theta):
    global mu
    v = np.sqrt(mu/a)
    dV = 2.0*v*np.sin(0.5*theta) # SMAD 9-78
    return dV

def combined(v_i,v_f,theta): # SMAD 9-77
    return np.sqrt(v_i**2 + v_f**2 - 2.0*v_i*v_f*np.cos(theta))
```

For the combined maneuvers, the initial velocity is our LEO circular orbit velocity, and the final velocity is the _LEO circular orbit velocity plus $\Delta V_L$_ from SMAD 9-69.
For the second burn, the initial velocity is our _GEO circular orbit velocity less the $\Delta V_H$_ from SMAD 6-70, and the final velocity is the GEO orbit velocity.

Now we can calculate and confirm the SMAD values.

```python
# Note that I only calculate the hohmann transfer velocities once!
# The magnitude of these are the same no matter inclination - where we use them is different!
[dV_L, dV_H] = hohmann(r_circ, r_geo)
# For the combined maneuvers, we need the circular orbit velocities
v_circ = np.sqrt(mu/r_circ)
v_geo  = np.sqrt(mu/r_geo)

# Simple plane change at circular orbit, then hohmann to GEO
row1 = plane_change(r_circ, i_circ) + dV_L + dV_H
# Hohmann to GEO then plane change at GEO orbit
row2 = dV_L + dV_H + plane_change(r_geo, i_circ)
# Combined maneuver, with the plane change happening at perigee
row3 = combined(v_circ, v_circ+dV_L, i_circ) + dV_H
# Combined maneuver, with the plane change happening at apogee
row4 = dV_L + combined(v_geo-dV_H, v_geo, i_circ)
# Combined maneuvers, with the plane change split between the two
row5 = combined(v_circ, v_circ+dV_L, np.deg2rad(4.0)) + combined(v_geo-dV_H, v_geo, np.deg2rad(24.5))

output = """<table>
<tr><th>Approach</th><th>Total dV (km/s)</th></tr>
<tr><th>Plane change at 250km followed by Hohmann Transfer to GEO</th><td>{0:0.3f}</td></tr>
<tr><th>Hohmann Transfer to GEO followed by Plane change</th><td>{1:0.3f}</td></tr>
<tr><th>Combined Maneuver, all plane change at Perigee</th><td>{2:0.3f}</td></tr>
<tr><th>Combined Maneuver, all plane change at Apogee</th><td>{3:0.3f}</td></tr>
<tr><th>Combined Maneuver, 4deg plane change at Perigee, 24.5deg at Apogee:</th><td>{4:0.3f}</td></tr>
</table>""".format(row1,row2,row3,row4,row5)
display(HTML(output))
```

#### (b) The ESA has a launchpad in French Guiana located at $5^\circ$N. If they launch in the lowest possible inclination to a circular orbit of 250km altitude, how much delta-V do they need for the transfer to equatorial geosynchronous orbit if they do the entire plane change (i) at apogee? (ii) at perigee?

For a launchpad located at $5^\circ$N we can launch into an inclination of $5^\circ$ which we expect will save us some $\Delta V$, and money!

Let's take a look at the changes for row 3 and 4:

```python
i_circ=np.deg2rad(5.0)
# Combined maneuver, with the plane change happening at perigee
row3 = combined(v_circ, v_circ+dV_L, i_circ) + dV_H
# Combined maneuver, with the plane change happening at apogee
row4 = dV_L + combined(v_geo-dV_H, v_geo, i_circ)
output = """<table>
<tr><th>Approach</th><th>Total dV (km/s)</th></tr>
<tr><th>Combined Maneuver, all plane change at Perigee</th><td>{0:0.3f}</td></tr>
<tr><th>Combined Maneuver, all plane change at Apogee</th><td>{1:0.3f}</td></tr>
</table>""".format(row3,row4)
display(HTML(output))
```

So we save about 0.34 km/s for the plane change at apogee, however the important thing to note is how although plane changes are quite expensive near LEO (we save a massive 2.5 km/s for the plane change at perigee), they get much more palatable the further away from Earth you get.

### Bonus

While we are at it, let's take a quick look at what the minimum $\Delta V$ we can obtain for the ESA launchpad for varying levels of plane changes at apogee and perigee.

```python
min_dV = 100000.0
i_a, i_p = 0, 0
i_circ=np.deg2rad(5.0)
for i in np.linspace(0.0,i_circ,10000):
    new_dV = combined(v_circ, v_circ+dV_L, i) + combined(v_geo-dV_H, v_geo, i_circ-i)
    if new_dV < min_dV:
        min_dV = new_dV
        i_p = np.rad2deg(i)
        i_a = np.rad2deg(i_circ-i)
print 'Lowest delta-V is {0:0.3f} km/s, performing {1:0.2f} deg of plane change at perigee and {2:0.2f} at apogee.'.format(min_dV,i_p,i_a)
```

Barely any savings beyond what we already had. But what about confirming SMAD's assertation that for a LEO transfer, $4^\circ$ followed by $24.5^\circ$ is the most effecient?

```python
min_dV = 100000.0
i_a, i_p = 0, 0
i_circ=np.deg2rad(28.5)
for i in np.linspace(0.0,i_circ,10000):
    new_dV = combined(v_circ, v_circ+dV_L, i) + combined(v_geo-dV_H, v_geo, i_circ-i)
    if new_dV < min_dV:
        min_dV = new_dV
        i_p = np.rad2deg(i)
        i_a = np.rad2deg(i_circ-i)
print 'Lowest delta-V is {0:0.3f} km/s, performing {1:0.2f} deg of plane change at perigee and {2:0.2f} at apogee.'.format(min_dV,i_p,i_a)
```

This actually surprised me, we can get a lower $\Delta V$ than SMAD proposed.
