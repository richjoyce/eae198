Using this graph, determine the following:

#### (a) Determine the average orbital decay rate in meters per orbit for the decay between the two reboots from July 2013 to August 2013 and from late October 2013 to December 2013.

The way I attacked this is to draw the line of the average decay for each period, then estimated the altitude at the beginning and end of August and November, respectively.
This way I have an exact number of days.
Then, figure out the average number of orbits of the ISS per day.

```python
# From graph
h_aug1  = 416.4 # km
h_aug31 = 414.9 # km
h_nov1  = 417.6 # km
h_nov30 = 415.0 # km

# Daily decay rates 
daily_decay1 = (h_aug1 - h_aug31) / 31
daily_decay2 = (h_nov1 - h_nov30) / 30

# ISS orbits per day
a = r_e + 415.0  # Just using an approximate average
n = np.sqrt(mu/a**3)
P = 2*np.pi/n
orbits_per_day = (24.0 * 60.0 * 60.0) / P

# Orbital decay rate meters per orbit
orbital_decay1 = (daily_decay1 / orbits_per_day)*1000.0
orbital_decay2 = (daily_decay2 / orbits_per_day)*1000.0

print "Orbital Decay Rates"
print "July 2013-August 2013:\t\t {0:0.1f} m/orbit".format(orbital_decay1)
print "October 2013-December 2013:\t {0:0.1f} m/orbit".format(orbital_decay2)
```

#### (c)  Perform an "order of magnitude" approximation for the density of the atmosphere for the latter decay period. Use $C_D=2$, but justify and cite any other values you use.

The key equation here is SMAD (9-34), $\Delta a_{rev}=-2\pi(C_DA/m)\rho a^2$, which describes the change in radius per orbit.
Rearranging for $\rho$, the atmospheric density:
$$\rho = \frac{-\Delta a_{rev}m}{2\pi C_DAa^2}$$
We just determined $\Delta a_{rev}$ in Part A, and $C_D$ is given as $2$.
The mass can be found from [NASA](http://www.nasa.gov/mission_pages/station/main/onthestation/facts_and_figures.html), and I use $m=419,455$kg.
Area is the hardest to estimate, but lets just simply estimate a rectangle using the truss length of $109m$.
The height is a bit harder to find, but the solar arrays are $73m$ tall, so lets just use 20% of that to account for the solar arrays not covering all of the ISS surface area, and their variability.

```python
# ISS specs
m = 419455      # kg, mass
length = 109    # m
height = 0.2*73 # m
A = length*height
C_D = 2.0
a = r_e + 415.0 # km, again just using an approximate average

rho = (orbital_decay2*m)/(2.0*np.pi*C_D*A*(a*1000)**2)
print "Approximate density: {0:0.2e} kg/m^3 or {1:0.2e} kg/km^3".format(rho,rho*1e9)
```

#### (d) How much $\Delta$V was required for the reboost procedure at the beginning of October 2013?

In this re-boost, the ISS was raised from circular orbits at approximately 415 km to 418.5 km altitude.
Since this is such a small change we can just take the difference of orbital speeds.

```python
a_1 = 415.0
a_2 = 418.5
r_1 = r_e + a_1
r_2 = r_e + a_2
v_1 = np.sqrt( 398600.4418 / r_1 )
v_2 = np.sqrt( 398600.4418 / r_2 )
print "Delta V acheived: {0:0.2f} m/s".format( (v_1-v_2)*1000 )
```
