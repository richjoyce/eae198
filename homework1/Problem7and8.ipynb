```python
# Set up notebook environment
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
from IPython.display import HTML
# Set up some constants
r_e = 6378.1       # km; radius of the earth
mu = 398600.4418   # km^3/s^2; gravitational parameter of earth
```

The next two problems use the Clohessy-Wiltshire equations:

\begin{align}
\ddot{x} &= 3n^2{x} + {2}{w}{\dot{y}}\\
\ddot{y} &= -2{n}{\dot{x}}\\
\ddot{z} &= -n^2{z}\\
\end{align}
And their closed form solution:

\begin{align}
x(t) &= 4{x_0} + \frac{2}{n}\dot{y}_0 + \frac{1}{n}{\dot{x}_0}{\sin{nt}} - (\frac{2}{n}{\dot{y}_0} + 3{x_0})\cos{nt}\\
y(t) &=   y_0  - \frac{2}{n}\dot{x}_0 - 3(2n{x_0} + \dot{y}_0)t + 2(\frac{2}{n}{\dot{y}_0} + 3{x}_0)\sin{nt} + \frac{2}{n}{\dot{x}_0}\cos{nt}\\
z(t) &= \frac{1}{n}{\dot{z}_0}\sin{nt} + z_0{\cos{nt}}
\end{align}

# Problem 7. Pinwheel

When learning or thinking about orbital dynamics we often use the pinwheel diagram to remind ourselves how moving in one direction will induce a coupled motion in another.

In this problem you will be asked to recreate the pinwheel diagram more accurately as a plot (using MATLAB, etc. is highly recommended).
Use the closed form solution of the Clohessy-Wiltshire, a circular target orbit at 400km altitude, and the following initial conditions:

\begin{equation*}
\begin{aligned}
x_0 &= -100\mathrm{m}\\
y_0 &= -100\mathrm{m}\\
v_0 &= 1.0\mathrm{m/s}\\
\end{aligned}
\end{equation*}

#### (a) Plot $x$ vs $y$ for $0<t\leq5$minutes for an initial velocity direction in $x$, $-x$, $y$, $-y$ on the same plot.

This is simply plotting the result of the C-W equations for the orbital (x-y) plane for varying initial velocity directions.
First I define a function that will solve and return the x and y arrays.
\begin{align}
x(t) &= 4{x_0} + \frac{2}{n}\dot{y}_0 + \frac{1}{n}{\dot{x}_0}{\sin{nt}} - (\frac{2}{n}{\dot{y}_0} + 3{x_0})\cos{nt}\\
y(t) &=   y_0  - \frac{2}{n}\dot{x}_0 - 3(2n{x_0} + \dot{y}_0)t + 2(\frac{2}{n}{\dot{y}_0} + 3{x}_0)\sin{nt} + \frac{2}{n}{\dot{x}_0}\cos{nt}
\end{align}

```python
def cw_initial_xy(x0,y0,xd0,yd0,n,t):
    x = 4.0*x0 + (2.0/n)*yd0 + (1.0/n)*xd0*np.sin(n*t) - ((2.0/n)*yd0+3.0*x0)*np.cos(n*t)
    y = y0 - (2.0/n)*xd0 - 3.0*(2.0*n*x0+yd0)*t + 2.0*((2.0/n)*yd0+3.0*x0)*np.sin(n*t) + (2.0/n)*xd0*np.cos(n*t)
    return (x, y)
```

Now set up the initial conditions we need, call the functions and plot the results.

```python
x0 = -100.0   # m
y0 = -100.0   # m
v0 =  1.0     # m/s
t = np.linspace(0.0, 5.0*60, 1000)
n = np.sqrt( mu / (r_e+400.0)**3 )

x_r,y_r = cw_initial_xy(x0,y0, v0,  0,n,t);
x_l,y_l = cw_initial_xy(x0,y0,-v0,  0,n,t);
x_u,y_u = cw_initial_xy(x0,y0,  0, v0,n,t);
x_d,y_d = cw_initial_xy(x0,y0,  0,-v0,n,t);

fig, ax = plt.subplots()
ax.plot(y_r,x_r,y_l,x_l,y_u,x_u,y_d,x_d)
ax.set_xlim(-400, 400)
ax.set_ylim(-400, 400)
ax.axhline(0, color='black')
ax.axvline(0, color='black')
ax.invert_xaxis()
ax.set_ylabel("x - Relative distance along radius from Earth")
ax.set_xlabel("y - Relative distance along velocity vector")
ax.set_title("Pinwheel")
ax.grid()
```

#### (b) Does the pinwheel change shape depending on where you are relative to the target spacecraft? (i.e. different quadrants) Justify why this is using the equations.

No, the quadrant does not matter for the direction and general shape of the coupling.
The simple answer is that the formulas for $x(t)$ and $y(t)$ are dominated by the $\dot{x_0}/n$ and $\dot{y_0}/n$ terms as opposed to the $x_0$ and $y_0$.
So long as $\dot{x_0} \gg nx_0$ (or the same for $\dot{y},y$), which in this case is true since $nx_0$ is only 0.11.


Investigating this further, for $\dot{x_0}=0$, $\dot{y_0}\neq0$ we see a coupling that moves us in $\pm x$  respectively.
The equations reduce to,
\begin{align}
x(t) &= 4x_0 + \frac{2}{n}\dot{y_0} - \left(\frac{2}{n}\dot{y_0} + 3x_0\right)\cos{nt} \\
\frac{dx}{dt} &= \left(\frac{2}{n}\dot{y_0} + 3{x_0}\right)\sin{nt}
\end{align}

So you can see that the direction of the coupling is in the direction of $\left(\frac{2}{n}{\dot{y}_0} + 3{x_0}\right)$.


It does change the magnitudes of the coupling, which we can show this in a plot (this was not required, but I couldn't resist).

```python
t = np.linspace(0.0, 5.0*60, 1000)
n = np.sqrt( mu / (r_e+400.0)**3 )
v0 = 1.0
fig, ax = plt.subplots()
disp = 300.0
for x0 in [-disp, disp]:
    for y0 in [-disp, disp]:
        [x_r,y_r] = cw_initial_xy(x0,y0, v0,  0,n,t);
        [x_l,y_l] = cw_initial_xy(x0,y0,-v0,  0,n,t);
        [x_u,y_u] = cw_initial_xy(x0,y0,  0, v0,n,t);
        [x_d,y_d] = cw_initial_xy(x0,y0,  0,-v0,n,t);
        ax.plot(y_r,x_r,'blue',y_l,x_l,'green',y_u,x_u,'red',y_d,x_d,'cyan')
ax.set_xlim(-2*disp, 2*disp)
ax.set_ylim(-2*disp, 2*disp)
ax.axhline(0, color='black')
ax.axvline(0, color='black')
ax.invert_xaxis()
ax.set_ylabel("x - Relative distance along radius from Earth")
ax.set_xlabel("y - Relative distance along velocity vector")
ax.set_title("Pinwheel in different quadrants")
ax.grid()
```

If we scale the displacements and initial velocities, we see that the quadrant does matter, and the pinwheel breaks down.

```python
t = np.linspace(0.0, 5.0*60, 1000)
n = np.sqrt( mu / (r_e+400.0)**3 )
v0 = 0.1
disp = 100.0
fig, ax = plt.subplots()
for x0 in [-disp, disp]:
    for y0 in [-disp, disp]:
        [x_r,y_r] = cw_initial_xy(x0,y0, v0,  0,n,t);
        [x_l,y_l] = cw_initial_xy(x0,y0,-v0,  0,n,t);
        [x_u,y_u] = cw_initial_xy(x0,y0,  0, v0,n,t);
        [x_d,y_d] = cw_initial_xy(x0,y0,  0,-v0,n,t);
        ax.plot(y_r,x_r,'blue',y_l,x_l,'green',y_u,x_u,'red',y_d,x_d,'cyan')
ax.set_xlim(-2*disp, 2*disp)
ax.set_ylim(-2*disp, 2*disp)
ax.axhline(0, color='black')
ax.axvline(0, color='black')
ax.invert_xaxis();
ax.set_ylabel("x - Relative distance along radius from Earth")
ax.set_xlabel("y - Relative distance along velocity vector")
ax.set_title("Pinwheel in different quadrants")
ax.grid()
```

### Problem 8. SAFER

Now you are an astronaut.
During a routine EVA, you become separated from the space station.
You must use SAFER to fly yourself back to the space station.
Assume that the space station is at $\{x,y,z\} = \{0,0,0\}$ in the LVLH reference frame, and 400km altitude.

Use the Clohessy-Wiltshire equations and their closed-form solution to calculate the required $\Delta{v}$ to return to the space station in exactly 10 minutes for the following initial conditions.
(You may assume that the $\Delta{v}$ impulse is instantaneous.)

#### (a) \begin{align} \begin{bmatrix} x_0\\ y_0\\ z_0\\ \dot{x}_0\\ \dot{y}_0\\ \dot{z}_0 \end{bmatrix} &= \begin{bmatrix} 10m\\ 50m\\ 10m\\ 0\\ 0\\ 0\\ \end{bmatrix} \end{align}

We are solving the Clohessy-Wiltshire equations for $\dot{x_0}$, $\dot{y_0}$ and $\dot{z_0}$.
We know $x_0,y_0,z_0$ and want $x(600)=y(600)=z(600)=0$.
First let's set up the variables.

```python
x_0 = 10.0
y_0 = 50.0
z_0 = 10.0
a = r_e + 400.0
n = np.sqrt(mu/a**3)
t = 600.0
print n
```

Let's take a look at what we need to solve.
\begin{align*}
x(t) &= 4{x_0} + \frac{2}{n}\dot{y}_0 + \frac{1}{n}{\dot{x}_0}{\sin{nt}} - (\frac{2}{n}{\dot{y}_0} + 3{x_0})\cos{nt}\\
y(t) &=   y_0  - \frac{2}{n}\dot{x}_0 - 3(2n{x_0} + \dot{y}_0)t + 2(\frac{2}{n}{\dot{y}_0} + 3{x}_0)\sin{nt} + \frac{2}{n}{\dot{x}_0}\cos{nt}\\
z(t) &= \frac{1}{n}{\dot{z}_0}\sin{nt} + z_0{\cos{nt}}
\end{align*}

Rearranging these equations for the condition $x(600)=y(600)=z(600)=0$, and collecting like terms we get:
\begin{align*}
(3\cos(nt) - 4)x_0 &= \left[\frac{\sin(nt)}{n}\right]&\dot{x_0} &+ \left[\frac{2}{n}(1-\cos(nt))\right]&\dot{y_0} \\
y_0 + 6(\sin(nt)-nt)x_0 &= \left[\frac{2}{n}(1-\cos(nt))\right]&\dot{x_0} &+ \left[3t - \frac{4}{n}\sin(nt)\right]&\dot{y_0}
\end{align*}
\begin{align*}
-nz_0\frac{\cos(nt)}{\sin(nt)} = \dot{z_0}
\end{align*}
The first two can be solved as a linear system for the velocities, plugging in our known initial position, $n$ and $t$.
The $z$ equation can be solved directly.

```python
# Left hand side of the x/y equations
b = np.array([(3*np.cos(n*t)-4)*x_0, y_0+6.0*x_0*(np.sin(n*t)-n*t)])
# Matrix of dot(x0) and dot(y0)
A = np.mat([ [(np.sin(n*t)/n), (2.0/n)*(1.0-np.cos(n*t))], [(2.0/n)*(1.0-np.cos(n*t)), 3.0*t-(4.0*np.sin(n*t)/n)] ])
# Solve Ax = b
[xd_0, yd_0] = np.linalg.solve(A,b)
# Solve the uncoupled z equation easily
zd_0 = -n*z_0*np.cos(n*t)/np.sin(n*t)
# Put them in a single array
dV = np.array([xd_0, yd_0, zd_0])
print "xd0: {0:.4f} m/s\t\t yd0: {1:.4f} m/s\t zd0: {2:.4f} m/s".format(*dV)
print "Total Delta V: {0:.4f} m/s".format(np.linalg.norm(dV))
```

#### (b) \begin{align} \begin{bmatrix} x_0\\ y_0\\ z_0\\ \dot{x}_0\\ \dot{y}_0\\ \dot{z}_0 \end{bmatrix} &= \begin{bmatrix} 10m\\ 50m\\ 10m\\ 0\\ 0\\ 0\\ \end{bmatrix} \end{align}

Now we are considering the case where the initial velocity is not 0, which means we need to just do a difference as all that matters is our $v_0$ after the burn (which is $t=0$) is what we found above.

```python
v_0 = np.array([-0.5, -1.0, -0.5])
dVb = dV-v_0

print "xd0: {0:.4f} m/s\t\t yd0: {1:.4f} m/s\t zd0: {2:.4f} m/s".format(*dVb)
print "Total Delta V: {0:.4f} m/s".format(np.linalg.norm(dVb))
```
