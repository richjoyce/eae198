#### (a) What is required to launch into a perfectly circular orbit? Is this feasible?

Circular orbit requires $e=0$,
$$0 = \left[\left(\frac{r_0v_0^2}{\mu}-1\right)^2\cos^2\beta_0+\sin^2\beta_0\right]^{1/2}$$
$$\left(\frac{r_0v_0^2}{\mu}-1\right)^2\cos^2\beta_0 = -\sin^2\beta_0$$

For the simple case of $\beta_0=0$, then $\sin\beta_0=0$ and $\cos\beta_0=1$, leaving simply the term in the brackets.
This means we must satisfy $r_0v_0^2=\mu$, which is a known condition of a circular orbit, meaning that if $\beta_0=0$, then $e=0$.

For $\beta_0\neq 0$ then we can plug in $r_0v_0^2=\mu$, which cancels out the left side of the equation.
We are left only to satisfy $0 = \sin\beta_0$, however this is valid only at $\beta_0=0,\pi,2\pi,\dots$.
$\beta_0=0$ is the only physically realistic answer, which we already obtained.

To answer whether this is feasible, we consider the relation for flight path angle of $\sin\beta=\dfrac{\dot{r_0}}{v_0}$.
If $\beta=0$, then $\dot{r_0}$ must also equal zero.
This means we must burn out with no altitude rate, and the exact velocity required for a circular orbit at the altitude you are at.
This is technically feasible, but incredibly hard.

#### (b) The launch of STS-126 ended at MECO (Main-Engine Cut Off) with burnout conditions as given below.     \begin{align}     h &= 344674\mathrm{ft}\\     \dot{h} &= 324\mathrm{fps}\\     v_{(intertial)} &= 25818\mathrm{fps}\\     \end{align}

##### (i) Determine $\nu_0$ and $e$ of the orbit.

Here we are plugging into our formulas from above, noting that flight path angle is $\sin\beta = \dfrac{\dot{h_0}}{v_0}$

```python
h_0ft = 344674.0 # ft; burnout altitude
hdot_0ft = 324.0 # ft/s; burnout altitude rate
v_0ft =  25818.0 # ft/s; burnout velocity
## Convert to meters
fttokm = 0.0003048   # ft to km conversion
h_0 = h_0ft * fttokm
hdot_0 = hdot_0ft * fttokm
v_0 = v_0ft * fttokm
```

```python
r_0 = r_e + h_0
beta = np.arcsin(hdot_0/v_0)
""" Store r*v^2/mu """
rv2_mu = r_0 * v_0**2 / mu
""" Calculate the orbital parameters """
nu = np.arctan( (rv2_mu*np.sin(beta)*np.cos(beta))/(rv2_mu*(np.cos(beta)**2)-1.0) )
e = np.sqrt( (rv2_mu-1)**2 * (np.cos(beta)**2) + (np.sin(beta)**2) )
print "Flight path angle:\t{0:0.2f} deg".format(np.rad2deg(beta))
print "Degrees since perigee:\t{0:0.1f} deg".format(np.rad2deg(nu))
print "Eccentricity:\t\t{0:0.3f}".format(e)
```

##### (ii) Determine the minimum $\Delta{V}$ required to circularize this orbit. When should we fire the engines?

The $\Delta V$ required to circularize is the difference between the circular orbit speed at the current altitude and the current orbital speed.
The minimum will be at either apogee or perigee.
At perigee you will have to burn against the direction of travel (lower your speed) and at apogee you burn in the direction of travel, raising your speed.
Thus the equations for each are:
$$\Delta V_p = V_p - \sqrt{\frac{\mu}{r_p}}$$
$$\Delta V_a = \sqrt{\frac{\mu}{r_a}} - V_a$$

Operationally, burning at apogee is preferred as you then insert yourself into the highest orbit.


A messy rearrangement of the vis-viva equation can give you the semi-major axis, $a$, directly from a certain $r$ and $v$,
$$a=\left( \frac{2}{r}-\frac{v^2}{\mu} \right)^{-1}$$
Then, radius at perigee and apogee are found from the eccentricity $r_p=a(1-\epsilon)$ and $r_a=a(1+\epsilon)$.

```python
a = 1.0/((2.0/r_0) - ((v_0**2) / mu))
r_a = a*(1.0+e)
r_p = a*(1.0-e)

# Orbital speeds at apogee and perigee
v_a = np.sqrt(mu*(1-e)/r_a)
v_p = np.sqrt(mu*(1+e)/r_p)

# Circular orbits at apogee and perigee
v_c_a = np.sqrt(mu/r_a)
v_c_p = np.sqrt(mu/r_p)

dV_a = v_c_a - v_a
dV_p = v_p - v_c_p
print "DeltaV at Perigee:\t{0:2.3f} km/s or {1:3.1f} fps or {2:2.2f} m/s".format(dV_p,dV_p/fttokm,dV_p*1000)
print "DeltaV at Apogee:\t{0:2.3f} km/s or {1:3.1f} fps or {2:2.2f} m/s".format(dV_a,dV_a/fttokm,dV_a*1000)
```

Therefore, the lowest $\Delta V$ is found at apogee, though in this case its quite a modest minimum.

### Bonus

We can also show that apogee is mathematically the minimum by plugging in the vis-viva equation to a generic formula.
$$\Delta V = \left| \sqrt{\frac{\mu}{r}} - \sqrt{\mu\left(\frac{2}{r}-\frac{1}{a}\right)} \right|$$
$$\Delta V = \sqrt{\frac{\mu}{r}}\sqrt{\left| 1 - 2 + \frac{r}{a} \right| }$$
$$\Delta V = \left| \sqrt{\frac{\mu}{r}}\sqrt{1 - 2 + \frac{r}{a}} \right|$$
$$\Delta V = \sqrt{\mu}\sqrt{\left| \frac{1}{a} - \frac{1}{r} \right| }$$
Evaluating this at the two ends of the orbit, perigee and apogee, we can see that 
$$\Delta V_a = \left| \frac{1}{a} - \frac{1}{r_a} \right| = \left| \frac{1}{a} - \frac{1}{a(1+\epsilon)} \right| = \left| \frac{\epsilon}{a(1+\epsilon)} \right| = \left| \frac{\epsilon}{r_a} \right|$$
$$\Delta V_p = \left| \frac{1}{a} - \frac{1}{r_p} \right| = \left| \frac{1}{a} - \frac{1}{a(1-\epsilon)} \right| = \left| \frac{\epsilon}{a(1-\epsilon)} \right| = \left| \frac{\epsilon}{r_p} \right|$$
Which tells us that $\Delta V$ is inversely porportional to radius, thus we want to circularize at the highest possible radius, or apogee.
